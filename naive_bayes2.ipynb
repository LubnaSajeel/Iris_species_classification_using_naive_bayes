To implement the naive_bayes algorithm,firstly we need to import pandas and numpy.Then convert to dataframe

```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

We have to find missing values in data.For that we use isna function

```python
df.isna().sum()
```

we have to separate the data into input and output labels in the name of 2 variables

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(x)
print(y)
```

we have to train the data,so using train_test_split function we separate the data into trained and tested data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

Using Standardscaler normalize the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
print(x_train)
print(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
result=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
print(result)
print(score)
```
